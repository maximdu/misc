```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```python
def parse_one_article(element):
    title = element.find('h3', class_='gs_rt').find('a').text
    n_cited = element.find('div', class_='gs_fl').find_all('a')[2].text
    year = element.find('div', class_='gs_a').text
    
    return (title, n_cited, year)
```

```python
def parse_one_page(soup):
    elements = soup.find_all('div', class_='gs_ri')
    
    df = pd.DataFrame([
        parse_one_article(element) 
        for element in elements
    ], columns=['title', 'n_cited', 'year'])
        
    sleep(2)
    
    return df
```

```python
def parse_pages(query, n_pages=10):
    
    parsed_pages = []
    for i in range(n_pages):
        page_query = query + f'&start={i*10}'
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36 Edge/18.19582'}
        r = requests.get(page_query, headers=headers)
    
        print(r.status_code)
        
        soup = BeautifulSoup(r.text)
        
        parsed_pages.append(
            parse_one_page(soup)
        )
    
    return pd.concat(parsed_pages)
```

```python
df = parse_pages("""https://scholar.google.com/scholar?&q=rcpsp&hl=ru&as_sdt=0,5&as_ylo=2013""", n_pages=25)
df['n_cited'] = df['n_cited'].str.extract('Цитируется: ([\d]+)')#.astype(int)
df['year'] = df['year'].str.extract('(\d\d\d\d) - ')#.astype(int)
df = df.sort_values('n_cited', ascending=False)
```

```python
df['n_cited'] = df['n_cited'].fillna(0).astype(int)
```

```python
df['year'] = df['year'].astype(int)
```

```python
df = df.sort_values(['n_cited', 'year'], ascending=[False, False])
```
